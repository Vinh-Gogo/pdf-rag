```python
import os
import glob

# Tìm tất cả các thư mục page_*
page_dirs = glob.glob('../data/markdown/page_*')

# Duyệt qua từng thư mục
for page_dir in page_dirs:
    # Tìm các file có tiền tố page_cleared trong thư mục
    cleared_files = glob.glob(os.path.join(page_dir, '*_cleared.md'))
    
    # Xóa từng file
    for file_path in cleared_files:
        try:
            os.remove(file_path)
            print(f"Đã xóa: {file_path}")
        except Exception as e:
            print(f"Lỗi khi xóa {file_path}: {e}")

print("Hoàn thành!")
```

```python
from markdown_it import MarkdownIt
from markdown_it.presets import commonmark
import re

def markdown_to_clean_text(md_text):
    # Cấu hình trình phân tích
    md = MarkdownIt("commonmark")
    
    # Chuyển sang HTML trước
    html = md.render(md_text)
    
    # Làm sạch HTML và chuyển sang text
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html, "html.parser")
    
    # Loại bỏ code block để text sạch hơn
    for code in soup.find_all(['code', 'pre']):
        code.decompose()
    
    return soup.get_text(separator=' ', strip=True)

# Sử dụng
text = """## **CÔNG BỐ THÔNG TIN**

| Chỉ<br>số                                                       | Tiêu<br>chuẩn<br>GRI                                                                                                                                                | Trang<br>báo<br>cáo | Loại<br>trừ | Lý<br>do<br>loại<br>trừ            |
|-----------------------------------------------------------------|---------------------------------------------------------------------------------------------------------------------------------------------------------------------|---------------------|-------------|------------------------------------|
| 204-1                                                           | Tỷ<br>lệ<br>chi<br>tiêu<br>cho<br>các<br>nhà<br>cung<br>cấp<br>địa<br>phương                                                                                        | 149                 | Không       |                                    |
| GRI<br>205<br>Chống<br>tham<br>nhũng<br>2021                    |                                                                                                                                                                     |                     |             |                                    |
| 205-1                                                           | Các<br>hoạt<br>động<br>được<br>đánh<br>giá<br>rủi<br>ro<br>liên<br>quan<br>đến<br>tham<br>nhũng                                                                     | 110-114             | Không       |                                    |
| 205-2                                                           | Truyền<br>thông<br>và<br>đào<br>tạo<br>về<br>các<br>chính<br>sách<br>và<br>quy<br>trình<br>phòng<br>chống<br>tham<br>nhũng                                          | 114                 | Không       |                                    |
| 205-3                                                           | Các<br>sự<br>cố<br>tham<br>nhũng<br>đã<br>được<br>xác<br>nhận<br>và<br>biện<br>pháp<br>khắc<br>phục                                                                 | 114                 | Không       |                                    |
| GRI<br>206<br>Hành<br>vi<br>cản<br>trở<br>cạnh<br>tranh<br>2021 |                                                                                                                                                                     |                     |             |                                    |
| 206-1                                                           | Các<br>hành<br>động<br>pháp<br>lý<br>đối<br>với<br>hành<br>vi<br>hạn<br>chế<br>cạnh<br>tranh,<br>chống<br>độc<br>quyền<br>và<br>hình<br>thức<br>độc<br>quyền<br>bán |                     | Có          | Chưa<br>áp<br>dụng                 |
| GRI<br>207<br>Thuế<br>2019                                      |                                                                                                                                                                     |                     |             |                                    |
| 207-1                                                           | Chiến<br>lược<br>về<br>thuế                                                                                                                                         |                     | Có          | Thông<br>tin<br>không<br>có<br>sẵn |
| 207-2                                                           | Quản<br>trị,<br>kiểm<br>soát<br>và<br>quản<br>lý<br>rủi<br>ro<br>về<br>thuế                                                                                         |                     | Có          | Thông<br>tin<br>không<br>có<br>sẵn |
| 207-3                                                           | Sự<br>tham<br>gia<br>của<br>các<br>bên<br>liên<br>quan<br>và<br>phương<br>pháp<br>quản<br>lý<br>các<br>mối<br>quan<br>ngại<br>liên<br>quan<br>đến<br>thuế           |                     | Có          | Thông<br>tin<br>không<br>có<br>sẵn |
| 207-4                                                           | Báo<br>cáo<br>theo<br>quốc<br>gia                                                                                                                                   |                     | Có          | Thông<br>tin<br>không<br>có<br>sẵn |
| GRI<br>301<br>Vật<br>liệu<br>2016                               |                                                                                                                                                                     |                     |             |                                    |
| 301-1                                                           | Vật<br>liệu<br>được<br>sử<br>dụng<br>theo<br>trọng<br>lượng<br>hoặc<br>khối<br>lượng                                                                                | 67                  | Không       |                                    |
| 301-2                                                           | Sử<br>dụng<br>nguyên<br>liệu<br>đầu<br>vào<br>đã<br>tái<br>chế                                                                                                      |                     | Có          | Thông<br>tin<br>không<br>có<br>sẵn |
| 301-3                                                           | Sản<br>phẩm<br>bị<br>thu<br>hồi<br>và<br>vật<br>liệu<br>đóng<br>gói<br>sản<br>phẩm                                                                                  |                     | Có          | Thông<br>tin<br>không<br>có<br>sẵn |"""
clean_text = markdown_to_clean_text(text)
print(clean_text)  # Tiêu đề Đây là văn bản đậm. Item 1 Item 2
```
